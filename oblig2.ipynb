```python
import numpy as np
import matplotlib.pyplot as plt

#Fant ut skriftlig at den deriverte har et nullpunkt mellom 0 og 2, vet at x-verdien til toppunktet til f er mellom der.
a = 0
b = 2

#Setter opp den deriverte med a og b verdiene
fa = np.arctan(a)-(4/((a**2)+1))
fb = np.arctan(b)-(4/((b**2)+1))

#Bruker halveringsmetoden for å komme så nærme svaret som mulig
while(b-a >1e-4):
    c = (a+b)/2
    fc = np.arctan(c)-(4/((c**2)+1))
    if(fa*fc < 0):
        b = c
    else:
        a = c

#Setter x-verdien til det jeg fant med halveringsmetoden
fx = (a+b)/2
#Finner y-verdien som hører til x-verdien
fy = np.e**(-fx/4)*np.arctan(fx)

#Bestemmer verdiene som trengs for å plotte f(x)
x = np.linspace(0,5,100)
y = np.e**(-x/4)*np.arctan(x)

#Plotter f(x)
plt.plot(x,y,'r')
#Plotter maksimalpunktet
plt.plot(fx,fy,'bo',label = f'x = {fx:.4f}' + f'   y = {fy:.4f}')
plt.axis([0,4,0,3])
plt.grid(visible=True)
plt.legend()
plt.show()
```
